```python
#start imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# load the data
pH_values = [5.35, 6.10, 6.86]
pH_error = [0.05, 0.03, 0.02]

#create the figure
fig = plt.figure(figsize=(14,8))
ax = plt.axes((0.1,0.1,0.5,0.8))

ax.tick_params(axis='x', direction='in', labelsize=30, width=3, length=10)
ax.tick_params(axis='y', direction='in', labelsize=30, width=3, length=10)

#plot the data
ax.errorbar(range(1, 4), pH_values, yerr=np.array(pH_error) * np.sqrt(8), fmt='o', color='black', ecolor='grey', markersize=10, capsize=5, capthick=2, elinewidth=2)

#create the x labels
ax.set_xticks(range(1, 4), [f'{i}' for i in ['Interior', 'Edge', 'Outside']], rotation=45, fontsize=30)

plt.yticks(fontsize=30)
ax.set_ylim(4,8)
ax.set_xlim(0.5, 3.5)
ax.set_ylabel("pH", fontsize=36)
#sex linewidth of the axes
ax.spines['bottom'].set_linewidth(2)  # Make the x-axis line thicker
ax.spines['left'].set_linewidth(2)    # Make the y-axis line thicker
ax.spines['top'].set_linewidth(2)  # Make the x-axis line thicker
ax.spines['right'].set_linewidth(2)    # Make the y-axis line thicker
plt.savefig("pH_plot.svg", dpi=300, bbox_inches='tight')
plt.show()
```
